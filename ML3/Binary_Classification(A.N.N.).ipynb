```python
# ----------------------------------
# STEP 1: Import Required Libraries
# ----------------------------------
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
# ------------------------
# STEP 2: Create Dataset
# -------------------------

# Features: [number_of_links, spam_words_count, email_length]
X = np.array([
    [1, 2, 100],
    [0, 1, 200],
    [3, 5, 50],
    [2, 4, 70],
    [0, 0, 300],
    [4, 6, 40]
], dtype=float)

# Labels: 0 = Not Spam, 1 = Spam
Y = np.array([0, 0, 1, 1, 0, 1], dtype=float)


# ---------------------------
# STEP 3: Feature Scaling 
# ---------------------------

# Neural networks perform better when data is normalized
# Scaling values between 0 and 1

X = X / np.max(X, axis=0)
```

```python
# ----------------------------------------
# STEP 4: Create Neural Network Model
# ----------------------------------------

model = keras.Sequential([
    
    # Input layer (explicit definition - best practice)
    keras.Input(shape=(3,)),
    
    # Hidden layer with 4 neurons (increased capacity)
    keras.layers.Dense(4, activation='relu'),
    
    # Output layer with 1 neuron (for binary classification)
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# -------------------------
# Step 5: Compile Model
# -------------------------
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)


# --------------------------
# Step 6: Train Model
# --------------------------
model.fit(X, Y, epochs=200, verbose=0)


# ---------------------------
# Step 7: Test Prediction
# ---------------------------

# New email: [links=2, spam_words=3, length=80]
test_email = np.array([[2, 3, 80]])

prediction = model.predict(test_email)

print("Spam Probability:", prediction)

# Convert to class
if prediction >= 0.5:
    print("Result: Spam")
else:
    print("Result: Not Spam")

```

```python
# STEP 0: Import Required Libraries

import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Step 2: Create dataset

X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
Y = np.array([0, 0, 0, 1, 1, 1])
```

```python
# Step 3: Create model

model = LogisticRegression()
```

```python
# Step 4: Train model

model.fit(X, Y)

# Step 5: Predict

prediction = model.predict([[3.5]])
```

```python
# Step 6: Probability

prob = model.predict_proba([[3.5]])
```

```python
# Step 7: Output

print("Predicted Class:", prediction)
print("Probability:", prob)
import matplotlib.pyplot as plt
```

```python
# Scatter plot

plt.scatter(X, Y, color='blue')


# Logistic curve

X_test = np.linspace(0, 7, 100).reshape(-1, 1)
Y_prob = model.predict_proba(X_test)[:,1]

plt.plot(X_test, Y_prob, color='red')

plt.xlabel("Study Hours")
plt.ylabel("Probability of Passing")
plt.show()
```

##### This plot represents a Logistic Regression model where the red sigmoid curve shows the predicted probability of passing as a function of study hours.
##### The blue points represent the actual observed data (0 = fail, 1 = pass). As study hours increase, the probability smoothly transitions from 0 to 1, indicating a higher likelihood of passing. The midpoint (~0.5) acts as the decision boundary for classification.
